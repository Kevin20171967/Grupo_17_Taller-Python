**Ejercicio 2**


```python
data_ciudades = {
    'Ciudad': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'],
    'Gentilicio': ['Limense', 'Arequipeño', 'Trujillano', 'Cusqueño', 'Chiclayano', 'Piurano', 'Iquiteño', 'Huancaino', 'Tacneño', 'Pucallpino'],
    'Provincia': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Maynas', 'Huancayo', 'Tacna', 'Coronel Portillo'],
    'Región': ['Lima', 'Arequipa', 'La Libertad', 'Cusco', 'Lambayeque', 'Piura', 'Loreto', 'Junín', 'Tacna', 'Ucayali'],
}
```

```python
nombres = pd.DataFrame(data_ciudades)
```

```python
nombres
```

**Ejercicio 3**

```python
cuadro_1 = pd.merge(nfo, nombres, on='Ciudad', how='inner')
```

```python
cuadro_1
```

**Ejercicio 4**

```python
min_densidad = cuadro_1['Densidad_poblacion_por_km2'].min()
ciudad_min_densidad = cuadro_1[cuadro_1['Densidad_poblacion_por_km2'] == min_densidad]['Ciudad'].values[0]
```

```python
max_densidad = cuadro_1['Densidad_poblacion_por_km2'].max()
ciudad_max_densidad = cuadro_1[cuadro_1['Densidad_poblacion_por_km2'] == max_densidad]['Ciudad'].values[0]
```

```python
ciudad_min_densidad
```

```python
ciudad_max_densidad
```

**Ejercicio 5**

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.bar(cuadro_1['Ciudad'], cuadro_1['Habitantes'], color='darkblue'),
ax.set_xlabel('Ciudad')
ax.set_ylabel('Cantidad de Habitantes')
ax.set_title('Número de Habitantes por Ciudad en Perú')
ax.tick_params(axis="x", rotation=45)
plt.tight_layout()
plt.show()
```

**Ejercicio 6**

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(cuadro_1['Altitud_m'], cuadro_1['Habitantes'], color='teal', edgecolor='darkblue')
ax.set_xlabel('Altitud (m)')
ax.set_ylabel('Cantidad de Habitantes')
ax.set_title('Relación entre Altitud y Número de Habitantes de las ciudades del Perú')
plt.show()
```
